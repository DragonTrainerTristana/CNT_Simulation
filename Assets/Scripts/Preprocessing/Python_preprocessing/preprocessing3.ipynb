```python
import pandas as pd
df = pd.read_csv("CollisionCheck_preprocessed3.csv",header=None)
```

```python
rows, cols = df.shape
```

```python
for i in range(rows):
    start = False
    final = False
    count = 0
    othervalue = False
    dead = False
    for j in range(cols):
        if df.iloc[i,j] == "-1":
            df.iloc[i,j] = "-1:1"   
       
        if df.iloc[i,j] == "-1:0":
            start = True
        if df.iloc[i,j] == "-1:1" or df.iloc[i,j] == "-1":
          
            final = True
        if df.iloc[i,j] != "-1:0" and (df.iloc[i,j] != "-1:1" or df.iloc[i,j] == "-1") and pd.notna(df.iloc[i,j]):
            count += 1
        if count > 1:
            othervalue = True
    if start == True and count > 0:
        pass
    elif final == True and count > 0:
        pass
    elif othervalue == True:
        pass
    else:
        dead = True
        #df.iloc[i,cols-1] = "-1:-1"
    for j in range(cols):
        if dead == True:
            if pd.isna(df.iloc[i,j]):
                df.iloc[i,j] = "-1:-1"
                break
            
        
df.to_csv("CollisionCheck_preprocessed4.csv", header=None, index=False,na_rep="None")
```
